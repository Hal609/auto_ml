```python
import pandas as pd
import matplotlib.pyplot as plt
from clean import clean_data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
import seaborn as sns

sns.set(
    { "figure.figsize": (17, 7) },
    style='ticks',
    palette=sns.color_palette("Set2"),
    color_codes=True,
    font_scale=0.8
)
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
auto = pd.read_csv('adverts.csv')
# auto = clean_data(auto)
```

```python
# Initialize the PowerTransformer
pt = PowerTransformer(method='yeo-johnson', standardize=False)

# Step 1: Apply PowerTransformer
transformed_data = pt.fit_transform(auto[['price', 'mileage']])

# Step 2: Create a new DataFrame with transformed data
auto[['price', 'mileage']] = transformed_data

# Visualize transformed distributions
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(auto['price'], kde=True, bins=30)
plt.title("Transformed Price Distribution")

plt.subplot(1, 2, 2)
sns.histplot(auto['mileage'], kde=True, bins=30)
plt.title("Transformed Mileage Distribution")
plt.tight_layout()
plt.show()
```
